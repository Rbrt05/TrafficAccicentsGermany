```python
import pandas as pd
import numpy as np
import re
```

## Load and Concatenate Data

```python
# Load Data
pd.set_option('display.max_columns', None)
d2018= pd.read_csv('/Users/robertkammerer/Ironhack/05_Week/AccidentAtlasGermany/Data/Unfallorte2018_LinRef.txt', sep= ";")
d2019= pd.read_csv('/Users/robertkammerer/Ironhack/05_Week/AccidentAtlasGermany/Data/Unfallorte2019_LinRef.txt',sep=";")
d2020= pd.read_csv('/Users/robertkammerer/Ironhack/05_Week/AccidentAtlasGermany/Data/Unfallorte2020_LinRef.csv',sep=";")
```

```python
print(d2019.shape)
print(d2018.shape)
print(d2020.shape)
```

```python
d2018 = d2018.rename(columns={'OBJECTID_1': 'OBJECTID', 'IstSonstig': 'IstSonstige'})

d2018['id_dist'] = d2018['OBJECTID'].astype(str)+"_"+d2018['UJAHR'].astype(str)

d2018.head()
```

```python
d2019['id_dist'] = d2019['OBJECTID'].astype(str)+"_"+d2019['UJAHR'].astype(str)

d2019.head()
```

```python
d2020['id_dist'] = d2020['OBJECTID'].astype(str)+"_"+d2020['UJAHR'].astype(str)

d2020.head()
```

```python
d2020 = d2020.drop('UIDENTSTLAE',axis=1)
```

```python
# Full data for 2018-2020
data = pd.concat([d2018,d2019,d2020],axis=0)
```

```python
data.head()
```

## Clean Column Names

```python
data.columns = [x.lower() for x in data.columns]
```

```python
columns = list(data.columns)

columns = [re.sub("^u","",x) for x in columns]
print(columns)

data.columns = columns 

data = data.rename(columns={'objectid': 'id', 'regbez': 'bezirk',
                      'typ1': 'typ','lichtverh': 'lichtverhältnis','istrad': 'fahrrad',
                      'istpkw': 'pkw','istfuss': 'fußgänger','istkrad': 'motorrad',
                      'istgkfz': 'lkw','istsonstige': 'sonstige','strzustand': 'straßenzustand',})

data.head()
```

## Convert encoded data back to string

```python
data_clean=data.copy()
```

```python
# clean weekday

def clean_weekday(x):
    if x == 1:
        return "sunday"
    elif x == 2:
        return "monday"
    elif x == 3:
        return "tuesday"
    elif x == 4:
        return "wednesday"
    elif x == 5:
        return "thursday"
    elif x == 6:
        return "friday"
    elif x == 7:
        return "saturday"
    
data_clean['wochentag'] = data_clean['wochentag'].apply(clean_weekday)
```

```python
# clean injury

def clean_injury(x):
    if x == 1:
        return "fatal accident"
    elif x == 2:
        return "seriously injured"
    elif x == 3:
        return "slightly injured"
    
data_clean['kategorie'] = data_clean['kategorie'].apply(clean_injury)
```

```python
# clean type of accident

def clean_art(x):
    if x == 1:
        return "Collision with another vehicle which starts, stops or is stationary"
    elif x == 2:
        return "Collision with another vehicle moving ahead or waiting"
    elif x == 3:
        return "Collision with another vehicle moving laterally in the same direction"
    elif x == 4:
        return "Collision with another oncoming vehicle"
    elif x == 5:
        return "Collision with another vehicle which turns into or crosses a road"
    elif x == 6:
        return "Collision between vehicle and pedestrian"
    elif x == 7:
        return "Collision with an obstacle in the carriageway"
    elif x == 8:
        return "Leaving the carriageway to the right"
    elif x == 9:
        return "Leaving the carriageway to the left"
    else:
        return "Other"
    
data_clean['art'] = data_clean['art'].apply(clean_art)
```

```python
# clean type of accident 2

def clean_type(x):
    if x == 1:
        return "Driving accident"
    elif x == 2:
        return "Accident caused by turning off the road"
    elif x == 3:
        return "Accident caused by turning into a road or by crossing it"
    elif x == 4:
        return "Accident caused by crossing the road "
    elif x == 5:
        return "Accident involving stationary"
    elif x == 6:
        return "Accident between vehicles moving along in carriageway"
    elif x == 7:
        return "Other"
    
data_clean['typ'] = data_clean['typ'].apply(clean_type)
```

```python
# Clean Light

def clean_light(x):
    if x == 0:
        return "daylight"
    elif x == 1:
        return "twilight"
    elif x == 2:
        return "darkness"
    
data_clean['lichtverhältnis'] = data_clean['lichtverhältnis'].apply(clean_light)
```

```python
# Clean Condition

def clean_condition(x):
    if x == 0:
        return "dry"
    elif x == 1:
        return "wet"
    elif x == 2:
        return "slippery_winter"
    
data_clean['straßenzustand'] = data_clean['straßenzustand'].apply(clean_condition)
```

```python
# State

def clean_state(x):
    if x == 1:
        return "Schleswig-Holstein"
    elif x == 2:
        return "Hamburg"
    elif x == 3:
        return "Niedersachsen"
    elif x == 4:
        return "Bremen"
    elif x == 5:
        return "Nordrhein-Westfalen"
    elif x == 6:
        return "Hessen"
    elif x == 7:
        return "Rheinland-Pfalz"
    elif x == 8:
        return "Baden-Württemberg"
    elif x == 9:
        return "Bayern"
    elif x == 10:
        return "Saarland"
    elif x == 11:
        return "Berlin"
    elif x == 12:
        return "Brandenburg"
    elif x == 13:
        return "Mecklenburg-Vorpommern"
    elif x == 14:
        return "Sachsen"
    elif x == 15:
        return "Sachsen-Anhalt"
    elif x == 16:
        return "Thüringen"

data_clean['land_id'] = data_clean['land']
data_clean['land'] = data_clean['land'].apply(clean_state)
```

```python
# Vehicle

def fahrrad(x):
        if x == 1:
            return "fahrrad "
        else:
            return ""

def pkw(x):
        if x == 1:
            return "pkw "
        else:
            return ""

def fussgänger(x):
        if x == 1:
            return "fußgänger "
        else:
            return ""

def motorrad(x):
        if x == 1:
            return "motorrad "
        else:
            return ""

def lkw(x):
        if x == 1:
            return "lkw "
        else:
            return ""

def sonstige(x):
        if x == 1:
            return "sonstige"
        else:
            return ""


data_clean['combinations'] = (data_clean['fahrrad'].apply(fahrrad)+
                                data_clean['pkw'].apply(pkw)+
                                   data_clean['fußgänger'].apply(fussgänger)+
                                     data_clean['motorrad'].apply(motorrad)+
                                       data_clean['lkw'].apply(lkw)+
                                         data_clean['sonstige'].apply(sonstige))
```
